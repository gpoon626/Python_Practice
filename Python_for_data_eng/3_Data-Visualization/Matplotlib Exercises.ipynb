```python
x = np.arange(0,100)
y = x*2
z = x**2
```

### Exercise 1
1. Create a figure object using plt.figure()
2. Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax
3. Plot (x,y) on that axes and set the labels and titles to match the plot below:

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```
