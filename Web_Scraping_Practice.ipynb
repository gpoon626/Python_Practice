# #Example Project - Working w/ Multiple Pages & Items

```python
base_url = 'http://books.toscrape.com/catalogue/page-{}.html'
```

```python
#fill in page number with .format()
res = requests.get(base_url.format('1'))
```

```python
#grab books from get request

soup = bs4.BeautifulSoup(res.text,"lxml")

soup.select(".product_pod")
```

```python
products = soup.select(".product_pod")

example= products[0]

type(example)
```

```python
example.attrs
```

```python
list(example.children) #html info on first product
```

```python
# we see class: "star-rating Three" -- the '.' after rating can be seen in page inspection

example.select('.star-rating.Three')
```

```python
example.select('a')
```

```python
example.select('a')[1]
```

```python
example.select('a')[1]['title']
```

```python
#Let's try getting all the 4 star titles!

four_star_titles = []

for n in range(1,51):
    
    scrape_url = base_url.format(n)
    res = requests.get(scrape_url)
    
    soup = bs4.BeautifulSoup(res.text, "lxml")
    books = soup.select(".product_pod")
    
    for book in books:
        if len(book.select('.star-rating.Four')) != 0:
            four_star_titles.append(book.select('a')[1]['title'])
```
